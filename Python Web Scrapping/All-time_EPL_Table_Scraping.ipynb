```python
# Importing the Beautiful Soup library for web scraping
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Premier_League_records_and_statistics'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
# Find the table to get data from
table = soup.find_all('table')[6]
table
```

```python
# Find the headings
headings = table.find_all('th')
headings
```

```python
# Get text and strip
table_headings = [heading.text.strip() for heading in headings]
table_headings
```

```python
# Put headings into a dataframe
import pandas as pd
df = pd.DataFrame(columns = table_headings)
df
```

```python
# Find table content (rows)
table_rows = table.find_all('tr')
table_rows
```

```python
# Get data, strip and put content into the dataframe
for row in table_rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
# Define a function to clean the values in "Since/Last App." and "Pts" columns
import re
def subtract(value):
    pattern = r'\[.*?\]' 
    return re.sub(pattern, '', value) #Remove square brackets and their contents, which act as footnote in this case
```

```python
# Replace column "Since/Last App." and "Pts" values with clean values 
for idx, value in enumerate(df["Since/Last App."]):
    subtracted_value = subtract(value)
    df.loc[idx, "Since/Last App."] = subtracted_value
for idx, value in enumerate(df["Pts"]):
    subtracted_value = subtract(value)
    df.loc[idx, "Pts"] = subtracted_value
```

```python
# Group columns for converting dtype
to_int = ["Pos.", "Seasons", "Pld", "W", "D", "L", "GF", "GA", "GD", "Pts", "1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "T4", "T7", "Relegated"]
to_float = ["PPG"]
to_str = ["Debut", "Since/Last App.", "BestPos."]
```

```python
for column in to_int:
    df[column] = df[column].astype(str).str.replace('−', '-', regex=False)  # Replace '−' with '-'
    df[column] = df[column].str.replace(",", "")                            # Remove commas to convert to int
    df[column] = df[column].replace('', '0')                                # Replace empty strings with 0
    df[column] = df[column].astype(int)                                     # Convert to int
```

```python
for column in to_float:
    df[column] = df[column].astype(float)   # Convert to float
```

```python
for column in to_str:
    df[column] = df[column].apply(lambda x: re.sub(r'–', '-', str(x)))  # Replace '−' with '-' 
```
